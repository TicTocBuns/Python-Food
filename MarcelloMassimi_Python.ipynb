```python
def cerca_ricette(ricette, categoria=None, tempo_max=None, ingrediente_richiesto=None):
    #Cerca e mostra le ricette disponibili dato uno o piu' parametri
    risultati = []
    for ricetta in ricette:
        trovata = True
        if categoria is not None and ricetta['categoria'] != categoria:
            trovata = False
        if tempo_max is not None and ricetta['tempo cottura'] > tempo_max:
            trovata = False
        if ingrediente_richiesto is not None and ingrediente_richiesto not in ricetta['ingredienti']:
            trovata = False
        if trovata:
            risultati.append(ricetta)
    if risultati:
        print("Di seguito le ricette disponibili->")
        stampa_ricette(risultati)
    else:
        print("Non sono presenti ricette con questi parametri")
    return risultati

def conta_ricette(ricette, categoria=None, tempo_max=None):
    #Genera statistiche sulla quantità di ricette dato uno o piu' parametri
    count = 0
    for ricetta in ricette:
        if categoria is not None and ricetta['categoria'] != categoria:
            continue
        if tempo_max is not None and ricetta['tempo cottura'] > tempo_max:
            continue
        count += 1
    return count

def stampa_ricette(ricette):
    #Visualizza il ricettario
    print("Ricettario:\n" + "-" * 20)
    for ricetta in ricette:
        print(f"{ricetta['nome']}")
        print("Ingredienti:")
        for ingrediente in ricetta['ingredienti']:
            print(f"- {ingrediente}")
        print(f"Tempo di cottura: {ricetta['tempo cottura']} minuti")
        print(f"La categoria del piatto è: {ricetta['categoria']}")
        print("-" * 20)
        # Linea di separazione tra le ricette

def aggiungi_ricetta(ricette, nome, ingredienti, tempo_cottura, categoria):
    #Registra ricette nel ricettario
    nuovo_ricetta = {
        "nome": nome,
        "ingredienti": ingredienti,
        "tempo cottura": tempo_cottura,
        "categoria": categoria
    }
    ricette.append(nuovo_ricetta)
    return ricette
# Main:
ricette = []
ricette = aggiungi_ricetta(ricette, "Pasta al sugo",
                           ["pasta", "sugo"], 10, "primo")
ricette = aggiungi_ricetta(ricette, "Carbonara",
                           ["pasta", "uova", "guanciale", "pepe", "pecorino"], 15, "primo")
ricette = aggiungi_ricetta(ricette, "Biscotti",
                           ["farina", "burro", "zucchero"], 60, "dolce")
stampa_ricette(ricette)
print(f"Totale ricette:{conta_ricette(ricette, categoria='primo')}") #Prova 1
print(f"Totale ricette:{conta_ricette(ricette, tempo_max=16)}") #Prova 2
print(f"Totale ricette:{conta_ricette(ricette, categoria='dolce', tempo_max=14)}") #Prova 3
                        #categoria='...' per vedere quante ricette per categoria
                        #tempo_max=... per trovare una ricetta con x minuti o meno
#Vari esempi per il comando di ricerca
#cerca_ricette(ricette, categoria='dolce')
#cerca_ricette(ricette, tempo_max=15)
#cerca_ricette(ricette, categoria='dolce', tempo_max=14)
#cerca_ricette(ricette, ingrediente_richiesto='burro')
cerca_ricette(ricette, categoria='primo', ingrediente_richiesto='uova')
```
